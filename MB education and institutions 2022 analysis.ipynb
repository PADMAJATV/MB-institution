```python
x =['Avg score of MPC','Avg score of BiPC']
y =[500,900]
plt.bar(x,y,color="#0F0000",width=0.4)
plt.xlabel("Student")
plt.ylabel("strength of students")
plt.title("Student score status")
plt.show()
```

```python
x =d['District']
y =d['Rating of the clg']
plt.bar(x,y,color="#0F0000")
plt.xlabel("District")
plt.ylabel("Rating")
plt.title("MB COLLEGES RATING")
plt.show()
```
